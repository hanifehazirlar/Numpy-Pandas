## Kütüphaneleri import edin

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(11, dtype="float")
```

```python
m
```

```python
c = 3 * 10 ** 8
```

```python
E = m * c ** 2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.figure(figsize = (4.3, 3))

plt.plot(m, E, lw=4, color="b")

plt.xlabel("M in g", fontsize=7)                  
plt.xticks(fontsize=8)        

plt.ylabel("E in j", fontsize=7)                  
plt.yticks(fontsize=8)  

plt.title("E = mc^2", fontsize=9)        
                                                       
plt.show()
```

```python
plt.figure(figsize = (5.7, 4))

plt.plot(m, E, color = "orange", linewidth=2)

plt.xlabel("M in g", fontsize=7)                     
plt.xticks(fontsize=8)    

plt.ylabel("E in j", fontsize=7)                     
plt.yticks(fontsize=8)     

plt.title("E = mc^2", fontsize=9)    

plt.show()                                           
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
m = np.arange(11, dtype = "float")
m
```

```python
c = 3 * 10 ** 8
```

```python
E = m * c ** 2
```

```python
plt.figure(figsize = (4.3, 3))

plt.semilogy(m, E, color = "green", linewidth=2)     # semilogy() used to turn y-axis into log scaling

plt.xlabel("M in g", fontsize=7)                     
plt.xticks(fontsize=8)   
plt.xlim([0, 10])   

plt.ylabel("E in j", fontsize=7)                     
plt.yticks(fontsize=8)   

plt.title("E = mc^2", fontsize=9, x=0.5, y=0.985)                    
plt.grid(which='both', axis="y")

plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.figure(figsize=(5.8, 4.5))

plt.plot(labels, july16_2007, lw=1)
plt.xticks(fontsize=7.5)

plt.plot(labels, july16_2020, lw=1)
plt.yticks(fontsize=7.5)

plt.legend(["july16_2007", "july16_2020"], fontsize=7, loc="right")  # legend labels can be defined in legend() as well
plt.show()
```

```python
plt.figure(figsize=(5.8, 4.2))

plt.plot(labels,july16_2007, label = "july16_2007", lw=1)
plt.xticks(fontsize=7.5)

plt.plot(labels,july16_2020, label = "july16_2020", lw=1)
plt.yticks(fontsize=7.5)

plt.legend(fontsize=7, bbox_to_anchor=[1.23, 0.53])
plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig, ax = plt.subplots(2, 1, figsize=(7.3, 5.5))

ax[0].plot(labels,july16_2007, label = "july16_2007", lw=2)
ax[0].set_title("july16_2007", fontsize=9)
ax[0].tick_params(axis='x', labelsize=7.5)
ax[0].tick_params(axis='y', labelsize=7.5)

ax[1].plot(labels,july16_2020, label = "july16_2020", lw=2)
ax[1].set_title("july16_2020", fontsize=9)
ax[1].tick_params(axis='x', labelsize=7.5)
ax[1].tick_params(axis='y', labelsize=7.5)

plt.tight_layout();                        
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, ax1 = plt.subplots(figsize=(8.7, 5.7))

# Create a second y-axis
ax2 = ax1.twinx()

ax1.plot(labels, july16_2007, color='blue', label='July 16, 2007', lw=2)
ax1.set_ylabel("July 16 2007", color = "b", fontsize = 14)

ax2.plot(labels, july16_2020, color='red', label='July 16, 2020', lw=2)
ax2.set_ylabel('July 16 2020', color = "r", fontsize = 14)

ax1.tick_params(axis='x', labelsize=7)
ax1.tick_params(axis='y', labelsize=7)
ax2.tick_params(axis='y', labelsize=7)

plt.show()
```
