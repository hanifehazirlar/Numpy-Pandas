## Kütüphaneleri import edin

```python
import numpy as np
import pandas as pd

import matplotlib as mpl  
import matplotlib.pyplot as plt  

import seaborn as sns 

%matplotlib inline 

# %matplotlib notebook   

import warnings;
warnings.filterwarnings("ignore")
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(0,11,dtype=float)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.plot(m, E, color="b", linewidth = 5)
plt.show()
```

```python
plt.plot(m, E, color="orange", linewidth = 3)
plt.show()
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
logE = np.log10(E)
logE
```

```python
plt.figure(figsize=(6,4))
plt.plot(m, logE, color='green', linewidth=2)
plt.grid(axis = 'y')
plt.xlabel("E in J")
plt.ylabel("M in g" )
plt.title(label = "E=mc^2")

plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
plt.plot(labels, july16_2007, color = "b", label ="july16_2007")
plt.plot(labels, july16_2020, color = "orange", label ="july16_2020")

plt.legend()        
plt.show()
```

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
plt.plot(labels, july16_2007, color = "b", label ="july16_2007")
plt.plot(labels, july16_2020, color = "orange", label ="july16_2020")

plt.legend(bbox_to_anchor=(0.8, 0.0, 0.6, 0.6))        
plt.show()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(labels, july16_2007)
ax[0].set_title("july16_2007")
ax[1].plot(labels,july16_2020)
ax[1].set_title("july16_2020")
plt.tight_layout()
plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]

fig, ax1 = plt.subplots(figsize = (8.4,5))

# sol ekseni biçimlendiriyoruz
ax1.set_ylabel('July 16 2007', color = 'blue', size=14)
ax1.plot(labels, july16_2007, color = 'blue', linewidth=2)

#Burada sağ ekseni tanımlıyoruz 
ax2 = ax1.twinx()

# sağ ekseni biçimlendiriyoruz
ax2.set_ylabel('July 16 2020', color = 'red', size=14)
ax2.plot(labels, july16_2020, color = 'red', linewidth=2)
 
plt.show()
```

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]

fig, ax1 = plt.subplots(figsize = (8.4,5))

# sol ekseni biçimlendiriyoruz
ax1.set_ylabel('July 16 2007', color = 'blue', size=14)
ax1.plot(labels, july16_2007, color = 'blue', linewidth=2)
ax1.spines["left"].set_color("blue")  # Keşke bu da çalısaydı :((

#Burada sağ ekseni tanımlıyoruz 
ax2 = ax1.twinx()

# sağ ekseni biçimlendiriyoruz
ax2.set_ylabel('July 16 2020', color = 'red', size=14)
ax2.plot(labels, july16_2020, color = 'red', linewidth=2)
ax2.spines['right'].set_color('red')


plt.suptitle("Mustafa Hocam Alkışımı Rica Edeyim :)", color="purple", fontsize = 16, y = 0.94)      
plt.show()
```
