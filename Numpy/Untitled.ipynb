# Matplotlib

```python
x = [0,1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]
plt.plot(x,y) #default boyutta ve sekilde grafik olusturur
plt.show()
```

```python
# plt.plot() --> cizgi(line) grafigi
# plt.scatter() --> sacinti grafigi
# plt.bar() --> 
# plt.histogram()
```

```python
# parametreler:
# color: çizginin rengini belirler
# linestyle: çizgi stilini belirler
# marker: noktalara denk gelen kesişimleri gösterir. "o" olabilir, "s" olabilir.
# s: noktaların büyüklüğü
# edgecolors: noktaları saran renk.
# alpha: saydamlık - opaklık. 1=opak, 0=saydam
# label: grafiğin etiketi
```

```python
x = [0,1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]
plt.figure(figsize = (4,2)) # grafigin boyutunu ayarlar ve bos bir cerceve olusturur
                            # bu yuzden plt.plot(x,y) grafik adindan once yazariz
plt.plot(x,y) 
# farkli turde grafik olusturmak istedigimde plot yerine istedigim grafigin ismini yaziyoruz
plt.show()
```

```python
x = [0,1, 2, 3, 4, 5, 6] # x = np.arange(7)
y = [0, 2, 4, 6, 8, 10, 12] # y =np.arange(0,13,2) 

plt.figure(figsize = (4,2)) # grafigin boyutunu ayarlar
plt.plot(x,y,color = "orange", marker ="^") 
plt.show()
```

```python
x = np.arange(7)
y = -x*3+5

plt.figure(figsize = (4,2)) # grafigin boyutunu ayarlar
plt.scatter(x,y,color = "r", marker ="*",alpha = 0.5) 
plt.show()
```

# Scatter Plot(Sacinti Grafigi)

```python
# Scatter Plot(Sacinti Grafigi)
#seaborn.scatterplot(data=None, *, x=None, y=None, hue=None, size=None, style=None,
# palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, 
# size_norm=None, markers=True, style_order=None, legend='auto', ax=None, **kwargs)
```

```python
x = np.arange(7)
y = -x * 3 + 5
y2 = x * 5 -8

plt.figure(figsize = (12,6))
plt.scatter(x,y,marker = "*", color = "black", s = 150, alpha = 0.5, label = "yıldız")
plt.plot(x,y2, label = "line", marker = "^")
plt.legend() # etiketleri gormek isteresek legend() methodunu kullanmaliyiz
plt.show()
```

```python
x = np.arange(100)
y1 = np.sin(x)
plt.plot(x,y1);
```

```python
x = np.arange(100)/10
y1 = np.sin(x)
y2 = np.cos(x)
plt.scatter(x,y1, color = "r", s= 40,alpha = 0.5, edgecolors = "b", linewidths=10, label="sin(x)")
plt.plot(x,y2,label = "cos(x)")
plt.legend() # legend() fonksiyonu plot() icine label="cos(x)" parametresi yazilmalidir
plt.show()
# burada ayni figure ustunde farkli grafikler cizdirdik
```

```python
# gercek data ile example
np.random.seed(0)
data = {
    "Year": np.arange(2010, 2021),
    "Sell": np.random.randint(100, 500, 11),
    "Payment": np.random.randint(50, 400, 11),
    "Reagon": np.random.choice(["North", "South", "East", "West"], size = 11)
}
df = pd.DataFrame(data)
df
```

```python
# Tek bir figure
plt.plot(df["Year"], df["Payment"],
        label= "Payments",
        marker = "o",
        linestyle = "--",
        markerfacecolor = "yellow",
        color = "r")
plt.plot(df["Yil"], df["Sell"],
        label= "Sells",
        marker = "*",
        linestyle = "--",
        markerfacecolor = "black",
        color = "b")
plt.title("Annual Sales and Costs")
plt.xlabel("TL")
plt.ylabel("")
plt.legend()
plt.show()
```
